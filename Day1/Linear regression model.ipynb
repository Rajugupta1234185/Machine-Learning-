```python
x=np.array([1,2,3,4,5])
y=np.array([2,3,4,5,6])
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('plottig data')
plt.grid(True)
plt.show()
```

```python
m=0.0
b=0.0
def predict(x,m,b):
    return m*x+b
    
def loss(ypredict,ytrue):
    return np.mean((ypredict-ytrue)**2)




```

```python
def gradient_descent(x,m,b,y,lr):
    n=len(x)
    ypred=predict(x,m,b)
    dm=(-2/n)*np.sum(x*(y-ypred))
    db=(-2/n)*np.sum(y-ypred)
    m -=lr*dm
    b -=lr*b
    return m,b
    
    
```

```python
epochs=100
lr=0.01
for epoch in range(epochs):
    ypred=predict(x,m,b)
    compute_loss=loss(ypred,y)
    m,b=gradient_descent(x,m,b,y,lr)
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss={compute_loss:.4f}, m={m:.4f}, b={b:.4f}")
    
```

```python
plt.scatter(x, y, label="Actual")
plt.plot(x, predict(x, m, b), color='red', label="Predicted Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```
